```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Ano-2009.csv', delimiter=';')
```

```python
data.head()
```

```python
df = data[data['txNomeParlamentar'] == 'JAIR BOLSONARO']
```

```python
df['vlrDocumento'].iloc[0][0]
```

```python
primeiros = np.zeros(10)
primeiros
```

```python
df.shape
```

```python
for i in range(df.shape[0]):
    try:
        add = int(df['vlrDocumento'].iloc[i][0])
    except ValueError:
        add = int(df['vlrDocumento'].iloc[i][1])
    primeiros[add] += 1
```

```python
primeiros
```

```python
primeiros = np.delete(primeiros, 0)
primeiros
```

```python
sum_p = primeiros.sum()
sum_p
```

```python
primeiros_b = np.zeros(9)
for i in range(9):
    primeiros_b[i] = primeiros[i]/sum_p
```

```python
primeiros_b[0]
```

```python
benford = np.zeros(9)
for i in range(9):
    benford[i]  = np.log10(i + 2) - np.log10(i + 1)
benford
```

# Plotando gráficos

```python
from matplotlib import pyplot as plt
%matplotlib inline
eixo_x = np.array([0,2,4,6,8,10,12,14,16])
plt.bar(eixo_x,primeiros_b)
plt.bar(eixo_x+1,benford)
```
